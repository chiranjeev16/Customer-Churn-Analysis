```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading Dataset
df = pd.read_csv('Customer Churn.csv')
df
```

```python
df.info()
```

```python
#Replacing the blanks and changing datatype
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
#Dealing with duplicates
df["customerID"].duplicated().sum()
```

```python
#Data Transformation

def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
#Bar Plot

ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Churned Customers", fontsize = 10)
plt.show()
```

```python
# Pie Chart

plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.3f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churned Customers by Gender", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Churn by SeniorCitizen", fontsize = 10)
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust figsize for better visualization

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])  # Customize colors if desired

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))  # Customize legend location

plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.histplot(x = "tenure", data = df, bins = 72, hue = 'Churn')
plt.title("Tenure Of Churn Customers")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title("Churn Customers by Contracts")
plt.show()
```

```python
# List of columns to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set the number of subplots
num_cols = len(columns)
num_rows = (num_cols + 2) // 3  # Arrange in a grid (3 columns)

# Create subplots
fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows * 4))

# Flatten axes array for easier indexing
axes = axes.flatten()

# Loop through columns and create count plots
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i], hue = df['Churn'])
    axes[i].set_title(column)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Sample DataFrame (replace with your actual DataFrame)
# df = pd.read_csv('your_data.csv')  # Uncomment and replace with your DataFrame
# Example data creation for demonstration
np.random.seed(0)
df = pd.DataFrame({
    'tenure': np.random.randint(1, 73, 1000),  # Random tenure values between 1 and 72
    'Churn': np.random.choice(['Yes', 'No'], 1000)  # Random churn values
})

# Create a contingency table
heatmap_data = pd.crosstab(df['Churn'], df['tenure'])

# Create a heatmap
plt.figure(figsize=(17, 5))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='coolwarm', cbar=True)

# Add title and labels
plt.title("Tenure Of Churn Customers")
plt.xlabel("Tenure")
plt.ylabel("Churn")

# Show the plot
plt.show()
```
